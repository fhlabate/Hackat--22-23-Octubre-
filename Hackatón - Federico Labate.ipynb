## Limpieza de datos
Existen 104 registros que no contienen información sobre el código y el nombre del distrito, código y nombre del barrio, código y nombre de la calle donde se produjo el accidente y su código postal.



Estos registros representan el **1.16%** de los datos, por lo que eliminarlos podría ser una opción óptima y simple sin afectar significativamente la calidad del análisis. De esta manera simplifico el proceso y mejoro la calidad de futuros modelos.
